```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# Reference the file where the CSV is located
weather_csv_path = "../WeatherPy/output_data/city_weather.csv"

# Import the data into a Pandas DataFrame
weather_df = pd.read_csv(weather_csv_path)
weather_df.pop("Unnamed: 0")

# Display dataframe
weather_df
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# Store latitude and longitude in locations
locations = weather_df[["Lat", "Lng"]]

# Store humidity for weight of the heatmap
humidity = weather_df["Humidity"]

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius=3)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Filter dataframe for ideal conditions
hotel_df = weather_df[(weather_df["Max Temp"] < 80) & (weather_df["Max Temp"] > 70) & (weather_df["Wind Speed"] < 10) & (weather_df["Cloudiness"] == 0)]

# Display dataframe
hotel_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Create empty list to store hotel names
hotel_names = []

# Store latitude and longitude in series
hotel_lat = hotel_df["Lat"]
hotel_lat = hotel_lat.reset_index(drop=True)
hotel_lng = hotel_df["Lng"]
hotel_lng = hotel_lng.reset_index(drop=True)

# Loop through and store hotel name for each city
for x in range(0, 8):
    
    # Store latitude and longitude as string
    lat_lng = f'{hotel_lat[x]},{hotel_lng[x]}'
    
    # Store parameters for search
    target_coordinates = lat_lng,
    target_search = "hotel"
    target_radius = 5000
    target_type = "lodging"

    # set up a parameters dictionary
    params = {
        "location": target_coordinates,
        "keyword": target_search,
        "radius": target_radius,
        "type": target_type,
        "key": g_key,
    }

    # Build URL
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # Run request and store in json
    response = requests.get(base_url, params)
    hotel_response = response.json()
    
    # Store name and append to hotel name list
    hotel = hotel_response["results"][0]["name"]
    hotel_names.append(hotel)

# Add new column to hotel dataframe
hotel_df["Hotel Name"] = hotel_names

# Display dataframe
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map

# Create heat layer
hotel_layer = gmaps.marker_layer(locations, hover_text= hotel_info)

# Add layer
fig.add_layer(hotel_layer)

# Display figure
fig
```
